**Reflection Questions**


Regression Task: It predicts a continuous numerical value (the median house price in dollars), not a discrete category.

Classification Task: It predicts a discrete label or category (one of three specific species of Iris flowers).

Numeric vs. Categorical: A numeric target represents quantities where mathematical operations matter (e.g., price, height). A categorical target represents discrete groups or classes (e.g., spam/not spam, color).

Data Splitting: To evaluate how well the model generalizes to completely new, unseen data, preventing overfitting.

MSE for Classification: Categorical targets (0 and 1) lack continuous distance meaning. MSE penalizes wrong predictions non-convexly here, leading to poor optimization.

Confusion Matrix: It details exactly where a classification model succeeded and which specific classes it confused (False Positives and False Negatives).

Accuracy Limits: Imbalanced datasets render accuracy misleading. If 99% of samples are negative, a model predicting "negative" always is 99% accurate but completely useless.

Features vs. Labels: Features are the input variables used to make a prediction. Labels are the actual targets or answers you want to predict.

Test Size Impact: Increasing test size to 0.3 gives a more robust evaluation but reduces training data, which can slightly lower performance on small datasets.

Logistic Regression vs. Decision Tree: Logistic Regression creates linear boundaries; Decision Trees split data into orthogonal boxes, handling non-linear combinations better.

Linear vs. Ridge: Ridge regression adds a penalty to prevent coefficients from growing too large, reducing overfitting compared to standard linear regression.

**Mini Project**

1. Problem Identification: This is a regression task. The goal is to predict a continuous quantitative measure of disease progression one year after baseline.

2. Features and LabelsFeatures (10): Age, Sex, Body Mass Index (BMI), Average Blood Pressure (BP), and six blood serum measurements.

Label: A numerical score indexing disease progression.

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Load and Split
diabetes = load_diabetes()
X_train, X_test, y_train, y_test = train_test_split(diabetes.data, diabetes.target, test_size=0.2, random_state=42)

# Train
model = LinearRegression().fit(X_train, y_train)
preds = model.predict(X_test)

# Evaluate
rmse = np.sqrt(mean_squared_error(y_test, preds))
r2 = r2_score(y_test, preds)

print(f"RMSE: {rmse:.2f}")
print(f"R2 Score: {r2:.2f}")
```

**Simple Explanation of Results**

RMSE interpretation: On average, our model's predictions of disease progression are off by about 53.48 points on the target scale.

R2 Score interpretation: The model explains roughly 45% of the variation in disease progression. It captures general trends but leaves room for improvement, likely because the biological relationships are more complex than a straight line.
