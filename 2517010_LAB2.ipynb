```python
import numpy as np
import scipy.linalg as la
from scipy import stats
import sympy as sp
import matplotlib.pyplot as plt

# --- 1. LINEAR ALGEBRA ---
# Task 1, 2, 4, & 5
A = np.array([[4, 11], [1, 7]])
B = np.array([[2, 0], [0, 2]])

# Output 1: Matrix Multiplication
mat_mul = np.dot(A, B)
# Output 2: Determinant
det_A = np.linalg.det(A)
# Output 3: Eigenvalues
eigenvals, _ = np.linalg.eig(A)

print("--- LINEAR ALGEBRA OUTPUTS ---")
print(f"1. Matrix Product (A*B):\n{mat_mul}")
print(f"2. Determinant of A: {det_A:.2f}")
print(f"3. Eigenvalues of A: {eigenvals}\n")

# --- 2. CALCULUS ---
# Task 1, 3, & 5
x_sym = sp.Symbol('x')
f_sym = x_sym**3 - 6*x_sym**2 + 9*x_sym

# Output 1: Derivative Calculation
deriv = sp.diff(f_sym, x_sym)
# Output 2: Critical Points (where derivative is 0)
crit_pts = sp.solve(deriv, x_sym)

# Output 3: Gradient Descent (Finding local minimum)
def grad_descent(start_x, lr, steps):
    x = start_x
    for _ in range(steps):
        grad = 3*x**2 - 12*x + 9 # Manual derivative of f_sym
        x -= lr * grad
    return x
min_x = grad_descent(start_x=4, lr=0.1, steps=20)

print("--- CALCULUS OUTPUTS ---")
print(f"1. Derivative of f(x): {deriv}")
print(f"2. Critical Points: {crit_pts}")
print(f"3. Local Min via Gradient Descent: {min_x:.4f}\n")

# --- 3. PROBABILITY THEORY ---
# Task 1, 3, & 4
np.random.seed(42)
data = np.random.normal(100, 15, 1000)

# Output 1: Basic Probability (Likelihood of being > 115)
prob_high = np.mean(data > 115)
# Output 2: Bayes Theorem (P(A|B) for a diagnostic test)
p_disease = 0.01
p_pos_given_disease = 0.99
p_pos_given_no_disease = 0.05
p_pos = (p_pos_given_disease * p_disease) + (p_pos_given_no_disease * (1 - p_disease))
post_prob = (p_pos_given_disease * p_disease) / p_pos
# Output 3: Distribution Statistics
skewness = stats.skew(data)

print("--- PROBABILITY OUTPUTS ---")
print(f"1. P(X > 115) in sample: {prob_high:.4f}")
print(f"2. Posterior Probability (Bayes): {post_prob:.4f}")
print(f"3. Sample Skewness: {skewness:.4f}\n")

# --- 4. ADVANCED STATISTICS ---
# Task 2, 4, & 5
group_a = np.random.normal(50, 5, 100)
group_b = np.random.normal(52, 5, 100)

# Output 1: T-Test (Hypothesis Testing)
t_stat, p_val = stats.ttest_ind(group_a, group_b)
# Output 2: Maximum Likelihood Estimate (MLE) for mean
mle_mean = np.sum(group_a) / len(group_a)
# Output 3: Bootstrap Confidence Interval (95%)
boot_means = [np.mean(np.random.choice(group_a, size=len(group_a))) for _ in range(1000)]
ci = np.percentile(boot_means, [2.5, 97.5])

print("--- ADVANCED STATISTICS OUTPUTS ---")
print(f"1. T-Test P-Value: {p_val:.4f}")
print(f"2. MLE Mean of Group A: {mle_mean:.4f}")
print(f"3. 95% Bootstrap CI for Mean: {ci}")

# Visual Task: Functions and Graphs
x_plot = np.linspace(0, 4, 100)
y_plot = [float(f_sym.subs(x_sym, val)) for val in x_plot]
plt.plot(x_plot, y_plot, label='f(x) = x^3 - 6x^2 + 9x')
plt.title("Calculus: Function Visualization")
plt.legend()
plt.show()
```
